```python
import os
import natsort
from IPython.display import Image, display

output_dir = 'splat_output'
image_files = []
for filename in os.listdir(output_dir):
    if filename.endswith('.png'):
        image_files.append(os.path.join(output_dir, filename))

# Sort the files naturally by the number in their filename
image_files_sorted = natsort.natsorted(image_files)

print(f"Displaying {len(image_files_sorted)} rendered images:")
for img_path in image_files_sorted:
    print(f"Displaying {os.path.basename(img_path)}")
    display(Image(filename=img_path))
```
